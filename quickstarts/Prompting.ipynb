```
'''

Para entnder la funcion de simbolos matematicos

> = mayor que
< = menor que
>= = mayor o igual que
<= = menor o igual que
== = igual que
!= = diferente que

'''

nombre = 'Luis'
nombre1 = 'angelo'
edad = 23
hobby = 'motos'
#_motos = 7

print('Buenas noches!')
print('Como te llamas?', nombre)
print('Que edad tienes?', edad)
print('Que hobby te gusta?', hobby)

if edad >= 18:
    print('Veo que ya eres adulto!')
else:
    print('Aun te falta algo por crecer')

if nombre == nombre1:
  print('Eres la misma persona!')
else:
  print('Te nombre no me es familiar')

```
